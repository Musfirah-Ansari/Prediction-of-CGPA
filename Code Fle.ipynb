# CLEANING OF THE DATA

```python
#Function for comparing two algoritms
def comparing_algo(modeltrain, modeltraining, modeltest, modeltesting):
    plt.figure(figsize = (4, 10))
    plt.subplot(1, 2, 1)
    plt.bar(modeltrain, modeltraining, width = 0.3)
    plt.xlabel('Algorithm Names', fontsize = 16)
    plt.ylabel('Train Accuracy', fontsize = 16)
    plt.title('Bar plot for Training Accuracy', fontsize = 16)
    
    plt.subplot(1, 2, 2)
    plt.bar(modeltest,modeltesting, width = 0.3)
    plt.xlabel('Algorithm Names', fontsize = 16)
    plt.ylabel('Test Accuracy', fontsize = 16)
    plt.title('Bar plot for Testing Accuracy', fontsize = 16)
    
    plt.subplots_adjust(right = 5)
    plt.show()
```

```python
#Plotting the barplot between 2 algorithm for the analysis of their comparision
model1train = np.array(['KNN Regressor train', 'Decision Tree Regression train']) #model train
model1test = np.array(['KNN Regressor test', 'Decision Tree Regression test']) #model test
model1_training = np.array([m1train, m2train]) #model training
model1_testing = np.array([m1train, m2test]) #model testing

comparing_algo(model1train, model1_training, model1test, model1_testing)
```
